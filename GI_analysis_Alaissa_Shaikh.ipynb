#### **Analysis of success status**
---



```python
# Create a new column for success status
data['Success'] = data['Application Status'].apply(lambda x: 1 if x == 'Registered' else 0)
```

```python
# Group by goods and calculate the success rate
success_rate = data.groupby('Goods').agg({'Success': ['sum', 'count']})
success_rate.columns = ['Successful Applications', 'Total Applications']
success_rate['Success Rate'] = (success_rate['Successful Applications'] / success_rate['Total Applications']) * 100
```

```python
# Sort the success rate for better visualization
success_rate_sorted = success_rate.sort_values(by='Success Rate', ascending=False)
```

```python
# Print the success rates
print("Success Rates by Goods Category:")
print(success_rate_sorted[['Successful Applications', 'Total Applications', 'Success Rate']])
```

```python
# Plotting the success rate
plt.figure(figsize=(12, 6))
plt.bar(success_rate_sorted.index, success_rate_sorted['Success Rate'], color='teal')
plt.xlabel('Goods Category')
plt.ylabel('Success Rate (%)')
plt.title('Success Rate of GI Applications by Goods Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*   **High Success Rates**: This suggests that applications in these categories are almost always successful, potentially due to well-defined characteristics or strong regional associations that meet GI criteria effectively.


*   **Moderate Success Rates**: Ranging around 50-70%. These categories likely represent a mix of straightforward cases and more complex ones where GI criteria might be harder to satisfy consistently.

*   **Lower Success Rates**: Between about 30% and 50%. This variation could be due to more stringent criteria, diverse product characteristics, or less well-documented traditional practices in these categories.


*   **Lowest Success Rates**: Below 30%. These categories may face challenges such as competitive markets with similar products, difficulties in establishing uniqueness, or rigorous standards for proving geographical specificity.

#### **Distribution of GI applications by goods category**


---





```python
goods_distribution = data['Goods'].value_counts()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=goods_distribution.index, y=goods_distribution.values, palette='viridis')
plt.title('Distribution of GI Applications by Goods')
plt.xticks(rotation=45)
plt.ylabel('Number of Applications')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')

plt.show()
```

*   **Cultural and Economic Significance**: The high numbers in Handicraft and Agriculture suggest these sectors are culturally and economically significant for the regions applying for GIs. Protecting these goods can help preserve cultural heritage and support local economies.

*   **Promotion and Support**: Goods categories with fewer applications might benefit from increased promotional efforts or support from governing bodies to raise awareness about the benefits of GI protection.

*   **Policy Focus**: Understanding which goods categories have the most applications can help policymakers focus resources and support where it is most needed or where it can have the most substantial impact.
